### **Content Delivery Network (CDN)**  

  

A **Content Delivery Network (CDN)** is a globally distributed network of **proxy servers**, serving content from locations **closer to the user**. Generally, **static files** such as **HTML, CSS, JavaScript, images, and videos** are served from a CDN to reduce latency and improve loading speed.  

**Example:** Amazon CloudFront  

---

### **Push CDN**  

A **Push CDN** receives new content **whenever** changes occur on the server. The **server takes full responsibility** for providing content, uploading it directly to the CDN.  

We can configure **when content expires and when it is updated**. Content is uploaded **only when it is new or changed**, minimizing traffic but maximizing storage.  

#### **When to use a Push CDN?**  
 Suitable for **sites with a small amount of traffic** or **sites with content that is not frequently updated**.  

---

### **Pull CDN**  

A **Pull CDN** fetches new content from the **server** when the **first user requests it** and **caches it** for subsequent users.  

#### **When to use a Pull CDN?**  
 Ideal for **sites with dynamic content** or **frequently updated content**.  

---

### **Time-to-Live (TTL)**  

**TTL** determines **how long content is cached** before it expires and needs to be refreshed.  

---

### **Disadvantages of CDNs**  

1.  CDN costs can be significant depending on traffic, but this should be compared to the additional costs of not using a CDN.  
2.  If content is updated before the **TTL expires**, outdated content might still be served.  
3.  CDNs require modifying URLs for static content to point to the CDN.
